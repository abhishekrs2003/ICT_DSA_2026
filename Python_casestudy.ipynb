```python
class BankAccount:

    def __init__(self, owner, acc_no, balance):
        self.owner = owner
        self.acc_no = acc_no
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposit successful! Current balance: {self.balance}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount")
        elif amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawal successful! Current balance: {self.balance}")
        else:
            print("Insufficient balance")

    def check_balance(self):
        print(f"Current balance: {self.balance}")

onwer_name=input("Enter the name of the owner :")
acc_number=int(input("Enter the account number :"))
acc_bal=0

owner1 = BankAccount(onwer_name, acc_number,acc_bal)


while True:
    print("Select operations")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")

    choice = input("Enter your choice: ")

    match choice:
        case '1':
            amt = float(input("Enter amount to deposit: "))
            owner1.deposit(amt)

        case '2':
            amt = float(input("Enter amount to withdraw: "))
            owner1.withdraw(amt)

        case '3':
            owner1.check_balance()

        case '4':
            print("Exiting... Thank you!")
            break

        case _:
            print("Invalid choice, try again")
```

```python
days=int(input("Enter the number of days you want to analise :"))
temperature=[]
humidity=[]
count=1

for i in range(days):
  temp=int(input("enter the temperatures in order :"))
  temperature.append(temp)

print()

for j in range(days):
  hum=int(input("enter the humidity in order :"))
  humidity.append(hum)

print()

print(f"temperatures of {days} days are {temperature} :")
print(f"Humidity of {days}  days are {humidity} :")
print()

def analyse_temp(temp,hum):
  if temp<30 and hum >10:
    return "Rainy"
  else:
    return "Sunny"


def average_temperature(t):
  avg_temp=sum(t)/len(t)
  return avg_temp


def average_humidity(h):
  avg_hum=sum(h)/len(h)
  return avg_hum

for i in range(len(temperature)):
  analysis=analyse_temp(temperature[i],humidity[i])
  print(f"The day {count} is {analysis}")
  count+=1

print()

avg_temp=average_temperature(temperature)
avg_hum=average_humidity(humidity)

print(f"Average temperature is {avg_temp}")
print(f"Average humidity is {avg_hum}")
```
