#Pandas

###Basic Expolration

```python
import pandas as pd
import numpy as np
```

```python
# loading dataset into pandas dataframe
df=pd.read_csv('auto-mpg.csv')
```

```python
# first 10 rows
df.head(10)
```

```python
# total number of rows and coloumns
df.shape
```

```python
# Summary statistics for numerical columns
df.describe()
```

###Filtering and Indexing


```python
# all cars manufactured in 1975 with a weight less than 3000.
# Return the DataFrame with selected columns: car_name, weight, and mpg.

df[(df['model year']>75) & (df['weight']<3000)][['car name','weight','mpg']]
```

###Handling Missing Data

```python
df.isna()
df.isna().sum()
```

```python
# Replace missing values in the horsepower column with the column's median.

df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df=df.fillna({'horsepower':df['horsepower'].median()})
df
```

###Data Transformation

```python
# Add a new column power_to_weight_ratio, calculated as horsepower / weight.
df['power_to_weight_ratio'] = df['horsepower']/df['weight']
df
```

### Group By

```python
# Group the cars by origin
origin_group=df.groupby('origin')
df
```

```python
# calculate the mean mpg for each group.
origin_group.mean('mpg')
```

###Sorting

```python
# Sort the DataFrame by mpg in descending order

sorted_by_mpg=df.sort_values('mpg', ascending=False)
print(sorted_by_mpg)
```

```python
# display the top 10 cars with the highest mpg.

sorted_by_mpg.head(10)
```

###Apply function

```python
# Create a new column performance_score using a custom function:
def performance_score(row):
    return row['mpg'] * row['acceleration'] / row['weight']

df['performance_score'] = df.apply(performance_score, axis=1)
df
```

###Visualization Preparation


```python
# Generate a summary DataFrame with: Average mpg, weight, and horsepower for each model_year.
summary_df = df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean().reset_index()
print(summary_df)
```

###Exporting Data

```python
#Save a subset of the data containing only mpg, cylinders, horsepower, and weight for cars with mpg > 30 into a CSV file named high_mpg_cars.csv.

high_mpg_cars = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
high_mpg_cars.to_csv('high_mpg_cars.csv', index=False)
```

### Finding Anomalies

```python
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['mpg'] < lower_bound) | (df['mpg'] > upper_bound)]

outliers_df = outliers[['car name', 'mpg', 'model year']]

print(outliers_df)
```

#Numpy

###Basic Array Operations


```python
# Convert the mpg column into a NumPy array and calculate:
mpg_list = df['mpg'].values
```

```python
# The mean, median, and standard deviation of mpg.

mean_mpg = np.mean(mpg_list)
median_mpg = np.median(mpg_list)
std_mpg = np.std(mpg_list)
```

```python
# The number of cars with mpg greater than 25.
count_above_25 = np.sum(mpg_list > 25)
```

```python
print("Mean:", mean_mpg)
print("Median:", median_mpg)
print("Standard Deviation:", std_mpg)
print("Number of cars with mpg > 25:", count_above_25)
```

###Filtering

```python
# Using NumPy, filter all cars with more than 6 cylinders.Return the corresponding car_name as a list
cylinders = df['cylinders'].values
car_names = df['car name'].values

filtered_cars = car_names[cylinders > 6]

result = filtered_cars.tolist()

result
```

###Statistical Analysis

```python
# Compute the 25th, 50th, and 75th percentiles of the weight column using NumPy.
weight_array = df['weight'].values

p25 = np.percentile(weight_array, 25)
p50 = np.percentile(weight_array, 50)
p75 = np.percentile(weight_array, 75)

print("25th percentile:", p25)
print("50th percentile (median):", p50)
print("75th percentile:", p75)
```

###Array Manipulation


```python
# Convert the acceleration column into a NumPy array and normalize its values
acc_array = df['acceleration'].values

acc_norm = (acc_array - np.min(acc_array)) / (np.max(acc_array) - np.min(acc_array))

print(acc_norm)
```

###Broadcasting


```python

hp = pd.to_numeric(df['horsepower'], errors='coerce').values

mean_hp = np.nanmean(hp)
hp = np.where(np.isnan(hp), mean_hp, hp)

hp_updated = hp * 1.10

hp_updated
```

### Boolean Indexing

```python
# Find the average displacement of cars with an origin of 2 (Europe) using NumPy indexing.
displacement = df['displacement'].values
origin = df['origin'].values

europe_disp = displacement[origin == 2]

avg_disp = np.mean(europe_disp)

print("Average displacement (Europe):", avg_disp)
```

### Matrix Operations


```python

hp = pd.to_numeric(df['horsepower'], errors='coerce').values

# Create a 2D NumPy array containing the columns mpg, horsepower, and weight.
hp = np.where(np.isnan(hp), np.nanmean(hp), hp)
X = np.column_stack((df['mpg'].values, hp, df['weight'].values))

v = np.array([1, 0.5, -0.2])
# Compute the dot product of this matrix with a given vector [1, 0.5, -0.2].
result = np.dot(X, v)

result
```

###Sorting


```python
#Use NumPy to sort the cars by model_year in descending order and display the first five car names.

model_year = df['model year'].values
car_names = df['car name'].values

sorted_indices = np.argsort(model_year)[::-1]
top_5_cars = car_names[sorted_indices][:5]

top_5_cars
```

###Correlation

```python
# Compute the Pearson correlation coefficient between mpg and weight using NumPy.
mpg = df['mpg'].values
weight = df['weight'].values

corr_matrix = np.corrcoef(mpg, weight)
corr = corr_matrix[0, 1]

print("Pearson correlation:", corr)
```

###Conditional Aggregates

```python
# Calculate the mean mpg for cars grouped by the number of cylinders using NumPy techniques.

mpg = df['mpg'].values
cylinders = df['cylinders'].values
unique_cyl = np.unique(cylinders)

for cyl in unique_cyl:
    mean_mpg = np.mean(mpg[cylinders == cyl])
    print(f"Cylinders: {cyl}, Mean MPG: {mean_mpg}")
```
